# GitHub Actions + Colab 测试 Notebook

这是一个简单的测试 Notebook，用于验证 GitHub Actions 自动执行 Colab 的流程。

```python
# 单元格 1: 打印欢迎信息
print("🚀 Hello from Colab!")
print("测试开始...")
```

```python
# 单元格 2: 检查 Python 环境
import sys
print(f"Python 版本：{sys.version}")
print(f"运行平台：{sys.platform}")
```

```python
# 单元格 3: 简单计算
import numpy as np

# 生成随机数据
data = np.random.randn(1000, 100)
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)

print(f"数据形状：{data.shape}")
print(f"均值：{mean[:5]}")
print(f"标准差：{std[:5]}")
```

```python
# 单元格 4: 保存结果到 Drive
from google.colab import drive
import os

# 挂载 Drive
drive.mount('/content/drive')

# 创建输出目录
output_dir = '/content/drive/MyDrive/colab_test_results'
os.makedirs(output_dir, exist_ok=True)

# 保存结果
result_file = os.path.join(output_dir, 'test_result.txt')
with open(result_file, 'w') as f:
    f.write("✅ GitHub Actions + Colab 测试成功！\n")
    f.write(f"Python 版本：{sys.version}\n")
    f.write(f"数据形状：{data.shape}\n")
    f.write(f"测试时间：2026-03-12\n")

print(f"✅ 结果已保存到：{result_file}")
```

```python
# 单元格 5: 完成
print("="*50)
print("✅ 测试完成！")
print("="*50)
print("请检查你的 Google Drive: /colab_test_results/test_result.txt")
```
